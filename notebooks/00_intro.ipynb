In this directory, I'm going to guide you through building an AI application. My goal in the end is to have an application that lets you chat with any *PDF* document by simply uploading the document to the knowledge base. I'll use [**LangChain**](https://js.langchain.com/v0.1/docs/get_started/) as the framework and open-source models, some of which will be stored and managed locally. The focus will be on a technique called [**Retrieval Augmented Generation (RAG)**](https://arxiv.org/abs/2005.11401) to create a chatbot capable of learning from an external knowledge base.

The best part? This chatbot will be mostly cost-free, though running advanced techniques might incur some costs. For now, you can stick to the basic RAG technique without worrying about expenses. Let's dive in and explore how to implement this AI chatbot using LangChain and RAG. Keep in mind that you can swap most components for alternatives and still achieve the desired functionality.

I will be using this [**PDF**](https://pressbooks.oer.hawaii.edu/humannutrition2/) on nutrition, but you can use virtually any PDF and it should work fine.
